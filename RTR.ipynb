```python
### change height from feet' inches" to inches (float)
condition = df['height'] != 'nan'
df = df[condition]

def parse_ht(ht):
    # format: 7' 0.0"
    ht_ = ht.split("' ")
    ft_ = float(ht_[0])
    in_ = float(ht_[1].replace("\"",""))
    return (12*ft_) + in_

df['height'] = df['height'].astype(str)
df['height'] = df['height'].str.split('"')
df['height'] = df['height'].apply(lambda x:parse_ht(x))
```
